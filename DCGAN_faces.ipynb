# DCGAN to celebrities dataset

```python
dim = 64
colors = 3
```

```python
d_sizes = {
    'conv_layers': [
      (64, 5, 2, False),
      (128, 5, 2, True),
      (256, 5, 2, True),
      (512, 5, 2, True)
    ],
    'dense_layers': [],
  }
```

```python
g_sizes = {
    'z': 100,
    'projection': 512,
    'bn_after_project': True,
    'conv_layers': [
      (256, 5, 2, True),
      (128, 5, 2, True),
      (64, 5, 2, True),
      (colors, 5, 2, False)
    ],
    'dense_layers': [],
    'output_activation': tf.tanh,
  }
```

```python
 gan = DCGAN(dim, colors, d_sizes, g_sizes)
```

```python
Y = util.get_celeb()
```

```python
gan.fit(Y)
```
